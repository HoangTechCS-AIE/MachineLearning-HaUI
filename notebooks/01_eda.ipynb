Khám phá dữ liệu cổ phiếu, minh hoạ chỉ báo kỹ thuật và hệ thống chấm điểm rule-based.
Yêu cầu: đã chạy `scripts/01_crawl.py` + `scripts/02_features.py` (hoặc có sẵn `data/processed/final_merged.csv`).


```python
import sys; sys.path.insert(0, '..')
import pandas as pd, matplotlib.pyplot as plt
from src.config import load_config
from src.data.scoring import compute_scores, top_n_by_month
cfg = load_config()
```

```python
df = pd.read_csv(cfg.abs_path(cfg.paths.merged_file), parse_dates=['timestamp'])
print(df.shape)
print('Số mã:', df['ticker'].nunique())
df.head()
```

## Giá & đường trung bình động (chọn 1 mã)


```python
ticker = df['ticker'].iloc[0]
g = df[df['ticker']==ticker].sort_values('timestamp')
plt.figure(figsize=(12,5))
plt.plot(g['timestamp'], g['close'], label='Close', color='black')
plt.plot(g['timestamp'], g['sma20'], label='SMA20')
plt.plot(g['timestamp'], g['sma50'], label='SMA50')
plt.title(f'{ticker} — Giá & SMA'); plt.legend(); plt.show()
```

## RSI & MACD


```python
fig, ax = plt.subplots(2,1, figsize=(12,7), sharex=True)
ax[0].plot(g['timestamp'], g['rsi14'], color='purple'); ax[0].axhline(70,ls='--',c='r'); ax[0].axhline(30,ls='--',c='g'); ax[0].set_title('RSI14')
ax[1].plot(g['timestamp'], g['macd'], label='MACD'); ax[1].plot(g['timestamp'], g['macd_signal'], label='Signal'); ax[1].axhline(0,c='gray'); ax[1].legend(); ax[1].set_title('MACD')
plt.tight_layout(); plt.show()
```

## Chấm điểm rule-based & Top-10 theo tháng (baseline phi-ML)


```python
scored = compute_scores(df)
month = scored['timestamp'].dt.strftime('%Y-%m').max()
print('Tháng:', month)
top_n_by_month(scored, month, top_n=10)
```
